```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Iris-setosa
# Iris-versicolor
# Iris-virginica


df=pd.read_csv('iris.csv')
print(df)

print("pie chart of the species with percentage")
plt.pie(df[df.columns[5]].value_counts(), labels=df[df.columns[5]].unique(), autopct='%1.1f%%', startangle=140)
plt.title('Species Distribution wit Percentage')
legend = plt.legend(title='Species', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()



plt.bar(df['SepalLengthCm'], df['SepalWidthCm'])
plt.ylabel('Sepal Length (cm)')
plt.xlabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python


# Iris-setosa
# Iris-versicolor
# Iris-virginica



# 1. Line Plot
plt.plot(df['SepalLengthCm'], df['SepalWidthCm'], color='blue', linestyle='-', marker='o', label='Sepal')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Line Plot: Sepal Length vs Sepal Width')
plt.legend()
plt.show()
# Parameters:
# - x, y: Data for x and y axes
# - color: Line color
# - linestyle: Line style ('-', '--', '-.', ':')
# - marker: Marker style for points ('o', 's', '^', etc.)
# - label: Label for legend

# 2. Bar Plot
plt.bar(df['Species'].unique(), df['Species'].value_counts(), color='green')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Plot: Species Count')
plt.show()
# Parameters:
# - x: Categories
# - height: Heights of bars
# - color: Bar color

# 3. Histogram
plt.hist(df['SepalLengthCm'], bins=10, color='purple', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram: Sepal Length Distribution')
plt.show()
# Parameters:
# - x: Data
# - bins: Number of bins
# - color: Bar color
# - edgecolor: Color of bar edges

# 4. Scatter Plot
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c='red', marker='x')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot: Sepal vs Petal Length')
plt.show()
# Parameters:
# - x, y: Data for axes
# - c: Color
# - marker: Marker style

# 5. Pie Chart
plt.pie(df['Species'].value_counts(), labels=df['Species'].unique(), autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart: Species Distribution')
plt.show()
# Parameters:
# - x: Data
# - labels: Labels for slices
# - autopct: Format for percentage
# - startangle: Starting angle

# 6. Box Plot
plt.boxplot([df['SepalLengthCm'], df['SepalWidthCm']], labels=['SepalLengthCm', 'SepalWidthCm'])
plt.title('Box Plot: Sepal Measurements')
plt.ylabel('cm')
plt.show()
# Parameters:
# - x: Data
# - labels: Labels for boxes

# 7. Area Plot (using fill_between)
plt.fill_between(df.index, df['SepalLengthCm'], color='skyblue', alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.title('Area Plot: Sepal Length')
plt.show()
# Parameters:
# - x, y: Data
# - color: Fill color
# - alpha: Transparency

# 8. Horizontal Bar Plot
plt.barh(df['Species'].unique(), df['Species'].value_counts(), color='orange')
plt.xlabel('Count')
plt.ylabel('Species')
plt.title('Horizontal Bar Plot: Species Count')
plt.show()
# Parameters:
# - y: Categories
# - width: Bar lengths
# - color: Bar color
```
